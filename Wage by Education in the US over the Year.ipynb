```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("wages_by_education.csv")
display (df.head())
```

```python
# Sort the DataFrame by the 'Year' column in ascending order
df = df.sort_values(by='year', ascending=True)

# Reset the index (optional, for clean indexing)
df = df.reset_index(drop=True)

# Display the sorted DataFrame
display (df.head())
```

```python
df.info()
```

```python
df_general = df.iloc[:, 0:6]
display (df_general.head())
```

```python
display (df_general.info())
display (df_general.describe())
```

```python
df_general.columns
```

```python
#visualize the general wage over the year

df_general = df_general.rename(columns={
    'year':'Year', 
    'less_than_hs' : 'Less than HS',
    'high_school': 'High School', 
    'some_college' : 'Some College', 
    'bachelors_degree' : 'Bachelors Degree', 
    'advanced_degree' : 'Advanced Degree'
})
```

```python
df_general.head()
```

```python
#Create interactive visual
import plotly.express as px
```

```python
fig = px.line(df_general, x="Year", y=df_general.columns,
              title='Average Wage Change Over the Year')

fig.show()
```

```python
df_general['Average Wage'] = df_general.iloc[:, 1:].mean(axis=1)

display (df_general.head())
```

```python
fig = px.line(df_general, x="Year", y="Average Wage",
              title='Average Wage Change Over the Year')

fig.show()
```

```python

# Calculate percentage change for each race and add it to the new DataFrame
df_general['Average Wage (% Change)'] = df_general['Average Wage'].pct_change() * 100


# Display the resulting DataFrame
display (df_general.head())
```

```python
fig = px.line(df_general, x="Year", y='Average Wage (% Change)',
              title='Average Wage %Change Over the Year')

fig.show()
```

```python
#Now we wanted to see the wage gap between gender
#First I wanted to take a look how the overall wage of Men

df_men = pd.concat([df.iloc[:, [0]], df.iloc[:, 6:11]], axis=1)
display (df_men.head())
display (df_men.describe())
```

```python
df_men.columns
```

```python
#Rename the Column

df_men = df_men.rename(columns={
    'year':'Year',
    'men_less_than_hs' : 'Men Less Than HS',
    'men_high_school' : 'Men High School',
    'men_some_college' : 'Men Some College',
    'men_bachelors_degree' : 'Men Bachelors Degree',
    'men_advanced_degree' : 'Men Advanced Degree'
})

display (df_men.head())
```

```python
#Second I wanted to take a look how the overall wage of Women

df_women = pd.concat([df.iloc[:, [0]], df.iloc[:, 11:16]], axis=1)
display (df_women.head())
display (df_women.describe())
```

```python
#Rename the Column

df_women = df_women.rename(columns={
    'year':'Year',
    'women_less_than_hs' : 'Women Less Than HS',
    'women_high_school' : 'Women High School',
    'women_some_college' : 'Women Some College',
    'women_bachelors_degree' : 'Women Bachelors Degree',
    'women_advanced_degree' : 'Women Advanced Degree'
})

display (df_women.head())
```

```python
fig_men = px.line(df_men, x="Year", y=df_men.columns,
              title="Average Men's Wage Change Over the Year")



fig_women = px.line(df_women, x="Year", y=df_women.columns,
              title="Average Women's Wage Change Over the Year")


fig_men.show()

fig_women.show()


```

```python
#As we can see it appears that even Women with Advanced Degree still have lower wage compare to Men with Bachelor Degree
gender_comparison = pd.DataFrame({
    'Year': df_men['Year'],  # Assuming Year is identical in both DataFrames
    'Men Bachelors Degree' : df_men['Men Bachelors Degree'],
    'Women Advanced Degree': df_women['Women Advanced Degree']
})


fig_comparison = px.line(gender_comparison, x="Year", y=gender_comparison.columns,
              title="Distinc Comparison Women vs Men Wage Change Over the Year")


fig_comparison.show()


```

```python
#And how the overall between Men and Women
# Ensure 'Year' is included in the calculation
# Add a new column for average wage for men and women in their respective DataFrames
df_men['average_men_wage'] = df_men.iloc[:, 1:].mean(axis=1)
df_women['average_women_wage'] = df_women.iloc[:, 1:].mean(axis=1)

# Create a new DataFrame combining the Year, men's average, and women's average
overall_gender = pd.DataFrame({
    'Year': df_men['Year'],  # Assuming Year is identical in both DataFrames
    'average_men_wage': df_men['average_men_wage'],
    'average_women_wage': df_women['average_women_wage']
})

# Optionally, calculate an overall average wage across genders
overall_gender['overall_average_wage'] = overall_gender[['average_men_wage', 'average_women_wage']].mean(axis=1)

# Display the resulting DataFrame
display (overall_gender.head())
display (overall_gender.describe())
```

```python
overall_gender.columns
```

```python
#Rename the Column

overall_gender = overall_gender.rename(columns={
   'average_men_wage' : 'Average Men Wage',
    'average_women_wage': 'Average Women Wage',
    'overall_average_wage' : 'Overall Average Wage'
})

display (overall_gender.head())
```

```python
fig_gender = px.line(overall_gender, x="Year", y=["Average Men Wage", "Average Women Wage"],
              title="Average Women vs Men Wage Change Over the Year")

fig_overall = px.line(overall_gender, x="Year", y="Overall Average Wage",
              title="Overall Average Wage Change Over the Year")


fig_overall.show()
fig_gender.show()
```

```python

# Select relevant race columns
gender_columns = ["Average Men Wage", "Average Women Wage", "Overall Average Wage"]

# Calculate percentage change for each race and add it to the new DataFrame
for gender in gender_columns:
    overall_gender[f'{gender} (% Change)'] = overall_gender[gender].pct_change() * 100


# Display the resulting DataFrame
display (overall_gender.head())
```

```python

fig_gender = px.line(overall_gender, x="Year", y=['Average Men Wage (% Change)','Average Women Wage (% Change)'],
              title="% Wage Change by Gender Change Over the Year")


fig_overall = px.line(overall_gender, x="Year", y=['Overall Average Wage (% Change)'],
              title="Overall % Wage Change Change Over the Year")

fig_overall.show()

fig_gender.show()
```

```python
#Now I wanna how the wage across the race and gender 
df_rg = pd.concat([df.iloc[:, [0]], df.iloc[:, 31:61]], axis=1)
display (df_rg.head())
display (df_rg.describe())
```

```python

# Melt the DataFrame to long format Year, race, gender, education
df_long = df_rg.melt(id_vars=["year"], var_name="attribute", value_name="value")

# Split the 'attribute' column into 'race', 'gender', and 'education'
df_long[['race', 'gender', 'education']] = df_long['attribute'].str.split('_', n=2, expand=True)

# Drop the original 'attribute' column
df_long = df_long.drop(columns=["attribute"])

# Rearrange columns for clarity
df_long = df_long[['year', 'gender', 'education', 'race', 'value']]

# Replace underscores with spaces in the 'education' column
df_long['education'] = df_long['education'].str.replace('_', ' ')


# Display the transformed DataFrame
display(df_long.head())
display (df_long.tail())

#Get csv data
df_long.to_csv("wage_by_education_US_tidy_data.csv", index=False)
```

```python
fig_race_gender = px.bar(df_long, x="race", y="value", color="gender", title="Wage gap by demographic")
fig_education_gender = px.bar(df_long, x="education", y="value", color="gender")

fig_race_gender.show()
fig_education_gender.show()
```
